```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv("student-score.csv")
print(df.head())
```

```python
# Display column names
print(df.columns)
```

```python
# Descriptive statistics
df.describe()
```

```python
# General information about the DataFrame
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

### Drop Unnamed Column

```python
# Drop the 'Unnamed: 0' column
df = df.drop("Unnamed: 0", axis= 1)
```

```python
print(df.info())
```

```python
# Clean 'WklyStudyHours' column
df["WklyStudyHours"]= df["WklyStudyHours"].str.replace("5 - 10","5 - 10")
df.head()
```

### gender Distribution

```python
# Gender Distribution
plt.figure(figsize = (5,5))
ax =sns.countplot(data = df, x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#### Form the above chart we have analysed that the number of female's in the data is more then the number of male's

```python
#gb : (GROUP BY)
# Group by 'ParentEduc' and analyze scores
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
sns.heatmap(gb, annot = True) # For displaying the value in the heat map
plt.title("Relationship Between Parents education and Student's Score")
plt.show()
```

 #### From the above chart we have concluded that the education of the parents have a good impact on their scores  

```python
#Group by 'ParentMaritalStatus' and analyze scores
gb2 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean',"WritingScore":'mean'})
gb2
```

```python
sns.heatmap(gb2, annot= True)
plt.title("Relationship Between Parents education and Student's Score")
plt.show()
```

#### From the above chart we have concluded that the ParentMaritalStatus have no effect or negligible impact

## Distribution of Ethnic Group

```python
# Pie chart for Ethnic Group distribution
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()


l = ["group A", "group B", "group C", "group D", "group E"]
gList= [groupA["EthnicGroup"], groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

print(gList)
plt.pie(gList, labels=l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Group's")
plt.show()

```
